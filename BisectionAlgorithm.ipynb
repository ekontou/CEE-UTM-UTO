# Bisection Method
We are going to leverage the bisection method/algorithm to come up with the minimum of the following objective function:

$f(x)=65x^6+71x^5-322x^4-401x^3$

s.t. $1\leq x \leq3$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 3

x = np.linspace(a, b)

def f(x):
    return 65*x**6+71*x**5-322*x**4-401*x**3
def df(x):
    return 65*6*np.power(x,5)+71*5*np.power(x,4)-322*4*np.power(x,3)-401*3*np.power(x,2)

plt.grid()
plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

Bisection method code: run it in-place many times (Ctrl-Enter)

```python
m = (a+b)/2 #interb

if np.sign(df(m)) < 0:
    a = m
else:
    b = m
        
print(a, b)
x = np.linspace(a, b)
plt.grid()
plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```
